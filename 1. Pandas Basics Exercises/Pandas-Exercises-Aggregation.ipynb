# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
occ=pd.read_clipboard(sep="|")
occ.head()
```

### Step 3. Assign it to a variable called users.

```python
users=occ
users.occupation.value_counts()
```

### Step 4. Discover what is the mean age per occupation

```python
grp_occ=users.groupby(["occupation"])
grp_occ["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users["male_count"]=np.where(users.gender=="M",1,0)
male_per=((users.groupby(["occupation"])["male_count"].sum()/users.occupation.value_counts())*100).round(2)
male_per=male_per.sort_values(ascending=False)
male_per
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
res=users.groupby(["occupation"]).age.aggregate(['min','max']).reset_index(drop=False)
res=res.rename(columns={'occupation':'Occupation','min':'Min Age','max':'Max Age'})
res
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age=users.groupby(by=['occupation','gender'])['age'].mean().round(2)
mean_age=mean_age.reset_index(drop=False)
mean_age=mean_age.rename(columns={'age':'mean'})
mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.head()
```

```python
users['female_count']=np.where(users.gender=='F',1,0)
users.head()
```

```python
per_f=((users.groupby(['occupation'])['female_count'].sum()/users.occupation.value_counts())*100).round(2)
per_m=((users.groupby(['occupation'])['male_count'].sum()/users.occupation.value_counts())*100).round(2)
final=pd.concat([per_f,per_m],axis=1).reset_index(drop=False)
final=final.rename(columns={'index':'Occupation',0:'%age of Females',1:'%age of Males'})
final
```
